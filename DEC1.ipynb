```python
import random

# Take inputs
characters = input("Enter characters (comma separated): ").split(",")
places = input("Enter places (comma separated): ").split(",")
actions = input("Enter actions (comma separated): ").split(",")

# Strip extra spaces
characters = [c.strip() for c in characters]
places = [p.strip() for p in places]
actions = [a.strip() for a in actions]

# Generate random sentence
sentence = f"{random.choice(characters)} {random.choice(actions)} in {random.choice(places)}."

print("\nRandom Sentence:")
print(sentence)
```
