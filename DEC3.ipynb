```python
# np_ex2_index_slice.py
import numpy as np

if __name__ == "__main__":
    arr = np.arange(1, 26).reshape(5, 5)
    print("Original array:\n", arr)

    third_row = arr[2, :]
    print("\n3rd row:\n", third_row)

    second_col = arr[:, 1]
    print("\n2nd column:\n", second_col)

    block_2x2_bottom_right = arr[-2:, -2:]
    print("\n2x2 block bottom-right:\n", block_2x2_bottom_right)

    evens = arr[arr % 2 == 0]
    print("\nAll even numbers:\n", evens)

    # Replace middle element (row 2, col 2, 0-indexed) with 999
    arr[2, 2] = 999
    print("\nAfter replacing middle element with 999:\n", arr)
```

```python
# np_ex3_reshaping.py
import numpy as np

if __name__ == "__main__":
    a = np.arange(1, 13).reshape(1, 12)
    print("Original 1x12:\n", a)
    a_3x4 = a.reshape(3, 4)
    print("\n3x4:\n", a_3x4)

    b = np.arange(1, 37).reshape(6, 6)
    print("\nOriginal 6x6:\n", b)
    b_3x12 = b.reshape(3, 12)
    b_2x18 = b.reshape(2, 18)
    print("\n3x12:\n", b_3x12)
    print("\n2x18:\n", b_2x18)

    flat = b.flatten()
    print("\nFlatten 6x6 to 1D:\n", flat)

    c = np.arange(1, 11)
    print("\nOriginal 1D:\n", c)
    col_vec = c.reshape(-1, 1)
    row_vec = c.reshape(1, -1)
    print("\nColumn vector:\n", col_vec)
    print("\nRow vector:\n", row_vec)
```

```python
# np_ex4_vectorization_vs_loops.py
import numpy as np

arr = np.arange(1, 100001)

def square_loop(a):
    out = []
    for x in a:
        out.append(x * x)
    return np.array(out)

def square_vectorized(a):
    return a ** 2

if __name__ == "__main__":
    print("Using %timeit, run inside an interactive shell or Jupyter.")
    print("Example:")
    print("%timeit square_loop(arr)")
    print("%timeit square_vectorized(arr)")
    print("Vectorized approach is usually faster.")
```

```python
# np_ex5_conditional_selection.py
import numpy as np

if __name__ == "__main__":
    marks = np.array([10, 45, 67, 89, 32, 56, 77, 90, 12])
    print("Marks:", marks)

    passing = marks[marks > 40]
    print("\nPassing marks (>40):", passing)

    between_50_80 = marks[(marks >= 50) & (marks <= 80)]
    print("Marks between 50 and 80:", between_50_80)

    marks_modified = marks.copy()
    marks_modified[marks_modified < 40] = 0
    print("After replacing marks <40 with 0:", marks_modified)

    passed_count = np.sum(marks > 40)
    print("Number of students passed:", passed_count)
```

```python
# pd_ex9_clean_students.py
import pandas as pd
import numpy as np
from io import StringIO

if __name__ == "__main__":

    # Embedded CSV dataset
    csv_data = """Name,Age,Marks,City
ram,20,87,Bangalore
sita,,91,chennai
john,???,abc,mumbai
ram,20,87,Bangalore
meera,21,,Hyderabad
,19,77,Delhi
kiran,22,85,
"""

    # Read CSV from the string
    df = pd.read_csv(StringIO(csv_data))
    print("Original:\n", df)

    print("\nMissing values per column:\n", df.isna().sum())

    # Replace "???" in Age with NaN
    df["Age"] = df["Age"].replace("???", np.nan)
    df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

    mean_age = df["Age"].mean()
    df["Age"] = df["Age"].fillna(mean_age)

    # Convert Marks to numeric
    df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")

    # Fill missing City
    df["City"] = df["City"].fillna("Unknown")

    # Drop duplicate rows
    df = df.drop_duplicates()

    # Clean Name column
    df["Name"] = df["Name"].fillna("Unknown")
    df["Name"] = df["Name"].str.title()

    print("\nCleaned DataFrame:\n", df)

    print("\nSummary statistics:\n", df.describe())

    # Top 3 highest scorers
    top3 = df.sort_values(by="Marks", ascending=False).head(3)
    print("\nTop 3 highest scorers:\n", top3)
```

```python
# pd_ex10_column_ops.py
import pandas as pd
import numpy as np
from io import StringIO

if __name__ == "__main__":

    # Embedded CSV dataset
    csv_data = """Name,Age,Marks,City
ram,20,87,Bangalore
sita,,91,chennai
john,???,abc,mumbai
ram,20,87,Bangalore
meera,21,,Hyderabad
,19,77,Delhi
kiran,22,85,
"""

    # Read CSV from string
    df = pd.read_csv(StringIO(csv_data))

    # Clean Age
    df["Age"] = df["Age"].replace("???", np.nan)
    df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
    df["Age"] = df["Age"].fillna(df["Age"].mean())

    # Clean Marks
    df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")
    df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

    # Add Pass/Fail column
    df["Pass/Fail"] = np.where(df["Marks"] >= 40, "Pass", "Fail")

    # Add Bonus Marks
    df["Bonus Marks"] = df["Marks"] + 5

    # Drop unnecessary columns
    df = df.drop(columns=["City"])   # remove if not needed

    # Rename columns
    df = df.rename(columns={"Marks": "Score"})

    # Convert Age to integer
    df["Age"] = df["Age"].astype(int)

    print(df)
```

```python
# task4_clean_dataset.py
import pandas as pd
import numpy as np
from io import StringIO

if __name__ == "__main__":

    # Embedded dirty dataset
    csv_data = """name,age,salary,city
ram,25,35000,Bangalore
sita,,52000,Chennai
john,45,,Mumbai
meera,28,45000,Hyderabad
ram,25,35000,Bangalore
kiran, 38 ,70000, Delhi
arjun,41,48000,
,19,77000,Chennai
"""

    # Load dataset
    df = pd.read_csv(StringIO(csv_data))

    # Remove duplicate rows
    df = df.drop_duplicates()

    # Strip spaces in string columns
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype(str).str.strip()

    # Fill missing values
    df = df.fillna({
        c: ("Unknown" if df[c].dtype == "object" else df[c].mean())
        for c in df.columns
    })

    # Convert all text to lowercase
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].str.lower()

    # Convert numeric columns if possible
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = pd.to_numeric(df[col], errors="ignore")

    print("Cleaned DataFrame:\n", df.head())
```

```python
# netflix_analyze.py
import pandas as pd

if __name__ == "__main__":
    df = pd.read_csv("netflix.csv")
    print("Data:\n", df)

    top_rated = df.loc[df["Rating"].idxmax()]
    print("\nTop-rated movie/series:\n", top_rated)

    print("\nMovies after 2015:\n", df[df["Year"] > 2015])

    print("\nAll Sci-Fi:\n", df[df["Genre"] == "Sci-Fi"])

    print("\nAverage rating:", df["Rating"].mean())

    print("\nMost common genre:\n", df["Genre"].mode()[0])

    def duration_to_int(x):
        try:
            return int(str(x).split()[0])
        except:
            return None

    df["Duration_min"] = df["Duration"].apply(duration_to_int)
    longest = df.loc[df["Duration_min"].idxmax()]
    print("\nLongest movie/series:\n", longest)

    print("\nGrouped by Type (Movie/Series):\n", df.groupby("Type")["Rating"].mean())
```

```python
# netflix_visuals.py
import pandas as pd
import matplotlib.pyplot as plt

if __name__ == "__main__":
    df = pd.read_csv("netflix.csv")

    # Bar plot: average rating by genre
    genre_rating = df.groupby("Genre")["Rating"].mean()
    genre_rating.plot(kind="bar")
    plt.title("Average Rating by Genre")
    plt.xlabel("Genre")
    plt.ylabel("Rating")
    plt.tight_layout()
    plt.show()

    # Count of movies per year
    year_counts = df["Year"].value_counts().sort_index()
    year_counts.plot(kind="bar")
    plt.title("Count of Titles per Year")
    plt.xlabel("Year")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

    # Histogram of IMDb ratings
    df["Rating"].plot(kind="hist", bins=5)
    plt.title("Histogram of Ratings")
    plt.xlabel("Rating")
    plt.tight_layout()
    plt.show()
```
